```python
"""
新建文件夹
"""
import os
os.makedirs('dataset',exist_ok=True)
os.makedirs('dataset/images', exist_ok=True)
os.makedirs('dataset/labels', exist_ok=True)
os.makedirs('dataset/images/train', exist_ok=True)
os.makedirs('dataset/images/val', exist_ok=True)
os.makedirs('dataset/labels/train', exist_ok=True)
os.makedirs('dataset/labels/val', exist_ok=True)
```

```python
"""
把图片保存到train里面
"""
import os
import time
from PIL import Image
import cv2
from tqdm import tqdm
root_path = "C:/Users/31136/yolo_CelebA_Spoof/CelebA_Spoof"
paths = os.walk(root_path)
def find_path():
    for path , dirs, files in paths:
        for file in tqdm(files):
            if file.endswith(".jpg"):
                #找到类别
                cata = path.replace("\\", "/").split("/")
                image = Image.open(os.path.join(path,file))
                x,y,w,h,p = open(os.path.join(path,file[:-4]+"_BB.txt").replace("\\", "/"),"r").read().split("\n")[0].split(" ")
                x,y,w,h = float(x)/224,float(y)/224,float(w)/224,float(h)/224
                #把人脸位置
                loc = f"{x+w/2} {y+h/2} {w} {h}"
                if cata[-1] == "live":
                    open(f"dataset/labels/train/{file[:-4]}.txt","w").write("1"+" "+loc)
                else:
                    open(f"dataset/labels/train/{file[:-4]}.txt","w").write("0"+" "+loc)
                image.save(f"dataset/images/train/{file}")
find_path()
```
